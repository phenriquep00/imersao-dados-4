```python
# Desafio 3
# Preço do metro quadrado por bairro e plotar em um gráfico ideal.

data_filtered = data_filtered[data_filtered['Bairro'].notnull()]

bairros = {
    'Bairros': data_filtered['Bairro'].unique(),
    'Preço_m2': float()
} # estrutura do dataframe com cada bairro

df_preco_m2 = pd.DataFrame(data=bairros)
preco_m2 = []
for bairro in df_preco_m2['Bairros']:
  metragem_total = sum(data_filtered.loc[data_filtered['Bairro'] == bairro, 'Metragem'])
  valor_total = sum(data_filtered.loc[data_filtered['Bairro'] == bairro, 'Valor_anuncio_float'])
  
  preco_m2.append(valor_total / metragem_total)

df_preco_m2['Preço_m2'] = preco_m2
df_preco_m2.head()
```
