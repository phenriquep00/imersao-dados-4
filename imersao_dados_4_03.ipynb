```python
# Desafio 1

bairros = data['Bairro'].values # Nome dos bairros.
dict_media_metragem = {}

for bairro in bairros:
  media_metragem = data.loc[data['Bairro'] == bairro, 'Metragem'].mean() # Calcula a média da metragem de acordo com o nome do bairro.
  dict_media_metragem[bairro] = media_metragem # Adiciona a média ao dicionário.

dict_media_metragem
```

```python
data_filtered['Valor_milhoes'] = data_filtered['Valor_anuncio_float']
data_filtered['Valor_milhoes'] = data_filtered['Valor_milhoes'] / 1000000
data_filtered.head()
```

```python
# Desafio 2
# Deixar o gráfico do histograma de valores legível (alterar labels, cores, título, escala).

sns.set()
f, ax = plt.subplots(figsize=(10,8))
plt.title('Imóveis a venda x Preço')
ax.ticklabel_format(style='plain', axis='both')
sns.histplot(data=data_filtered, x='Valor_milhoes', kde=True, color='gray')
plt.xlim(0, 10)

ax.set_xlabel("Preço (milhoes R$)")
ax.set_ylabel("Imoveis à venda")

plt.show()
```

```python
# Desafio 3
# Preço do metro quadrado por bairro e plotar em um gráfico ideal.

data_filtered = data_filtered[data_filtered['Bairro'].notnull()]

bairros = {
    'Bairros': data_filtered['Bairro'].unique(),
    'Preço_m2': float()
} # estrutura do dataframe com cada bairro

df_preco_m2 = pd.DataFrame(data=bairros)
preco_m2 = []
for bairro in df_preco_m2['Bairros']:
  metragem_total = sum(data_filtered.loc[data_filtered['Bairro'] == bairro, 'Metragem'])
  valor_total = sum(data_filtered.loc[data_filtered['Bairro'] == bairro, 'Valor_anuncio_float'])
  
  preco_m2.append(valor_total / metragem_total)

df_preco_m2['Preço_m2'] = preco_m2
df_preco_m2.head()
```
